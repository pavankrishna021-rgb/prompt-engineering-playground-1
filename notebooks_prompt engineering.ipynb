```python
# Prompt Engineering Playground
# Interactive comparison of prompting techniques
print("ðŸš€ Welcome to Prompt Engineering Playground!")
print("Techniques: Zero-shot | Few-shot | Chain-of-Thought")
```

```python
def zero_shot_example():
    prompt = "classify the sentiment: 'this movie was fantastic!'"
    response = "positive" # simulated LLM response
    return f"**zero_shot prompt:** {prompt}\n**response:** {response}"

print(zero_shot_example())    
```

```python
def few_shot_example():
    few_shot_prompt = """
examples:
text: 'love this product!' positive
text: 'waste of time' negative
text: 'pretty decent' neutral

classify: 'service was ok'"""
    response = "neutral"
    return f"**few_shot prompt:**\n{few_shot_prompt}\n**response:** {response}"

print(few_shot_example())    
```

```python
def chain_of_thought_example():
    cot_prompt = """Q: If a shirt costs $20 and goes on 25% discount, what's the sale price?

Step 1: Calculate discount amount
Step 2: Subtract from original price  
Step 3: Final answer:"""
    response = "$15"
    return f"*Chain-of-Thought:\n{cot_prompt}\nResponse:* {response}"

print(chain_of_thought_example())
```

```python
def compare_prompts(user_input):
    """Compare all 3 techniques on same input"""
    return f"""
Input: {user_input}

ðŸ”¹ *Zero-shot*: Classify sentiment â†’ Positive/Negative/Neutral
ðŸ”¸ *Few-shot*: Uses 2-3 examples â†’ More consistent  
ðŸ”º *Chain-of-Thought*: Step-by-step reasoning â†’ Best for math/logic

*Try:* 'The restaurant service took 2 hours for cold food'
"""

# Test it
print(compare_prompts("Your test input here"))
```

```python
# performance comparison table
results ="""
| Technique | Use case | Accuracy |
|-----------|----------|----------|
| Zero-shot | Simple tasks | 70% |
| Few-shot  | Formatting | 85% |
| CoT       | Complex reasoning | 94% |
"""
print(results)
```
